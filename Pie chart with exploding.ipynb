```python
#Pie Charts with exploding in Python
import matplotlib.pyplot as pyplot
```

```python
labels = ('Python', 'Java', 'Scala', 'C#')
```

```python
sizes= [45, 30, 15, 10]
```

```python
#only "explodes" the 1st slice (i.e. 'Python')
explode= (0.1, 0, 0, 0)
```

```python
pyplot.pie(sizes,
explode=explode,
labels=labels,
autopct='%1.f%%',
shadow=True,
counterclock=False,
startangle=45)
pyplot.show()
```
